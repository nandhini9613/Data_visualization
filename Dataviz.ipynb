```python

# Step 1: Import Libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Step 2: Load Dataset
# Assuming your dataset is already saved as CSV. You can also save it manually as CSV.
df = pd.read_csv('hotel_bookings.csv')  # <-- Save your Excel to CSV first!

# Step 3: Basic Overview
print(df.head())
print(df.info())
print(df.describe())

# Step 4: Data Cleaning (Optional based on need)
# Example: Fill missing values if necessary
df['children'].fillna(0, inplace=True)
df['country'].fillna('Unknown', inplace=True)

# Step 5: Visualizations

# 5.1 Booking Status: Canceled vs Not
cancel_counts = df['is_canceled'].value_counts()
plt.figure(figsize=(6,6))
plt.pie(cancel_counts, labels=['Not Canceled', 'Canceled'], autopct='%1.1f%%', startangle=90, colors=['#4CAF50','#FF5722'])
plt.title('Booking Cancellation Status')
plt.show()

# 5.2 Hotel Type: Resort Hotel vs City Hotel
hotel_counts = df['hotel'].value_counts()
sns.barplot(x=hotel_counts.index, y=hotel_counts.values, palette='coolwarm')
plt.title('Hotel Type Distribution')
plt.ylabel('Number of Bookings')
plt.show()

# 5.3 Meal Type Distribution
meal_counts = df['meal'].value_counts()
sns.barplot(x=meal_counts.index, y=meal_counts.values, palette='viridis')
plt.title('Meal Package Distribution')
plt.ylabel('Number of Bookings')
plt.show()

# 5.4 Top 10 Countries by Bookings
top_countries = df['country'].value_counts().head(10)
sns.barplot(x=top_countries.values, y=top_countries.index, palette='magma')
plt.title('Top 10 Countries with Most Bookings')
plt.xlabel('Number of Bookings')
plt.show()

# 5.5 People Composition (Adults, Children, Babies)
people = df[['adults', 'children', 'babies']].sum()
sns.barplot(x=people.index, y=people.values, palette='pastel')
plt.title('Total Count of Adults, Children, and Babies')
plt.ylabel('Total People')
plt.show()

# Step 6: Export to Excel (optional)
summary = {
    'Booking Status': cancel_counts,
    'Hotel Type': hotel_counts,
    'Meal Types': meal_counts,
    'Top Countries': top_countries
}


summary_df = pd.DataFrame(dict([(k, pd.Series(v)) for k,v in summary.items()]))

summary_df.to_excel('hotel_booking_summary.xlsx', index=False)

print("Summary exported successfully to Excel!")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Step 1: Load the dataset
df = pd.read_csv("hotel_bookings.csv")  # Make sure your file name is correct

# Step 2: Prepare data summaries
cancel_counts = df['is_canceled'].value_counts()
hotel_counts = df['hotel'].value_counts()
meal_counts = df['meal'].value_counts()
top_countries = df['country'].value_counts().head(10)

# Step 3: Create plots and save them as images
sns.set(style="whitegrid")

# 1. Booking Cancellation Plot
plt.figure(figsize=(6, 4))
sns.barplot(x=cancel_counts.index, y=cancel_counts.values, palette="Set2")
plt.title('Booking Cancellation Status')
plt.xlabel('Canceled (0 = No, 1 = Yes)')
plt.ylabel('Count')
plt.savefig('cancel_plot.png')
plt.close()

# 2. Hotel Type Plot
plt.figure(figsize=(6, 4))
sns.barplot(x=hotel_counts.index, y=hotel_counts.values, palette="pastel")
plt.title('Hotel Type Distribution')
plt.ylabel('Count')
plt.savefig('hotel_plot.png')
plt.close()

# 3. Meal Type Plot
plt.figure(figsize=(6, 4))
sns.barplot(x=meal_counts.index, y=meal_counts.values, palette="muted")
plt.title('Meal Type Preferences')
plt.ylabel('Count')
plt.savefig('meal_plot.png')
plt.close()

# 4. Top Countries Plot
plt.figure(figsize=(8, 5))
sns.barplot(x=top_countries.values, y=top_countries.index, palette="Blues_d")
plt.title('Top 10 Booking Countries')
plt.xlabel('Number of Bookings')
plt.savefig('country_plot.png')
plt.close()

# Step 4: Export charts and summaries to Excel using XlsxWriter
with pd.ExcelWriter('hotel_booking_visual_summary.xlsx', engine='xlsxwriter') as writer:
    # Write raw summary tables
    cancel_counts.to_frame(name='Count').to_excel(writer, sheet_name='Booking Status')
    hotel_counts.to_frame(name='Count').to_excel(writer, sheet_name='Hotel Types')
    meal_counts.to_frame(name='Count').to_excel(writer, sheet_name='Meal Types')
    top_countries.to_frame(name='Bookings').to_excel(writer, sheet_name='Top Countries')

    # Access the workbook and sheets
    workbook = writer.book

    # Insert plots into each sheet
    for sheet_name, img in zip(
        ['Booking Status', 'Hotel Types', 'Meal Types', 'Top Countries'],
        ['cancel_plot.png', 'hotel_plot.png', 'meal_plot.png', 'country_plot.png']
    ):
        worksheet = writer.sheets[sheet_name]
        worksheet.insert_image('D2', img)

print("✅ Excel file with visual charts created: hotel_booking_visual_summary.xlsx")
```
